## Quadratic Function and matplotlib

We want to find the roots of quadratic function $f(x) = a x^2 + b x + c$

This formula gives us the roots

$$
x_1, x_2 = \cfrac{-b \pm \sqrt{b^2 - 4ac}}{2a}
$$

Write a function that return the roots. Then plot the quadratic function and highlight the roots.

```python
def quad_roots(a, b, c):
    D = b**2 - 4*a*c
    x1 = (-b + D**.5)/(2*a)
    x2 = (-b - D**.5)/(2*a)
    return x1, x2
```

```python
print(quad_roots(1, -2, 5))
```

```python
print(type(quad_roots(1, -4, 3)[0]))

# print(quad_roots(1, -2, 5))      (class: complex, duck typing)
```

```python
import numpy as np


def quad_complexroots(a, b, c):
    a = np.complex(a, 0)
    b = np.complex(b, 0)
    c = np.complex(c, 0)
    D = b**2 - 4*a*c
    x1 = (-b + np.sqrt(D))/(2*a)
    x2 = (-b - np.sqrt(D))/(2*a)
    return x1, x2
```

```python
print(quad_complexroots(1, -2, 5))
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0, 4, 201)

def quad_fun(a, b, c, x):
    return a*x**2+b*x+c

a = 1
b = -4
c = 3
plt.plot(x, quad_fun(a, b, c, x))
plt.xlabel('x', fontsize = 18)
plt.ylabel('y', fontsize = 18)
plt.grid()
x1, x2 = quad_roots(a, b, c)
plt.plot(x1,quad_fun(a, b, c, x1),'.r', markersize = 14) # .r (marker+color)
plt.plot(x2,quad_fun(a, b, c, x2),'.r', markersize = 14)
plt.show()
```

## Finding the intersection of two lines


Let's consider two lines expressed as $y = m_1 x + b_1$ and $y = m_2 x + b_2$. We can find the intersection of these two lines by finding $x^*$ and $y^*$ that solves the system of equations below

$$
y^* = m_1 x^* + b_1, \\
y^* = m_2 x^* + b_2
$$

It is better to write in a matrix form:

$$
A 
\begin{bmatrix}
x^* \\
y^*
\end{bmatrix}
= b
$$


And solve for 

$$
\begin{bmatrix}
x^* \\
y^*
\end{bmatrix}
= A^{-1}b
$$

```python
import numpy as np
b1 = 0
m1 = 1.    # use dot, can be directly use A**(-1), but the result is different from matrix inverse
b2 = -1.
m2 = 2.

A = np.array([[-m1, 1],[-m2, 1]])
b = np.array([[b1],[b2]]) # ndarray (2, 1) [b1, b2]: (2,)
#print(np.shape(np.array([b1, b2])))

Ainv = np.linalg.inv(A)
#print(A**(-1))
#print(Ainv)

xy_ans = np.matmul(Ainv,b)


print(Ainv@b)
```
